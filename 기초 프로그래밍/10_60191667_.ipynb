```python
import random

class customer:
    def __init__(self,id,high):
        self.start = random.randint(1,high)
        while True :
            self.des = random.randint(1,high)
            if self.des != self.start :
                break
        self.id = id
        self.inelevator = False
        self.isfinished = False

class elevator:
    def __init__(self,high,customers):
        self.cur_floor = 1
        self.num_of_floor = high
        self.direction = 'up'
        self.register_list = customers[:]
        self.inelevator =[]
    def move(self):
        if self.direction == 'up':
            print('eleavtor :',self.cur_floor,'->',self.cur_floor+1)
            self.cur_floor += 1
        else :
            print('eleavtor :',self.cur_floor,'->',self.cur_floor-1)
            self.cur_floor -= 1
    def register_customer(self):
        for person in self.register_list[:]:
            if person.start == self.cur_floor :
                print('customer',str(person.id),' ->   Elevator')
                person.inelevator = True
                self.inelevator.append(person)
                self.register_list.remove(person)
    def cancel_customer(self):
        for person in self.inelevator[:] :
            if person.des == self.cur_floor :
                print('customer',str(person.id),' ->   Destination Floor')
                person.inelevator = False
                person.isfinished = True
                self.inelevator.remove(person)    
    
class building :
    def __init__(self):
        while True :
            try:
                self.building_h = int(input('Enter the Floor : '))
                if self.building_h > 1 :
                    break
                print('Wrong input')
            except :
                print('Wrong input')
        while True :
            try:
                self.num_of_customers = int(input('Enter the Number of customer : '))
                if self.num_of_customers > 0 :
                    break
                print('Wrong input')
            except :
                print('Wrong input')
        self.customers = []
        for i in range(self.num_of_customers):
            a = customer(i+1,self.building_h)
            self.customers.append(a)
        self.building_elevator = elevator(self.building_h,self.customers)
    def output(self):
        print("\nElevator's Current Floor :",self.building_elevator.cur_floor,
              "\tElevator's direction :",self.building_elevator.direction)
        print('Customer:')
        for person in self.customers :
            print('\tCustomer '+str(person.id)+'\t'+'current floor : '
                  +str(person.start)+'\t'+'destination floor : '+str(person.des))
            print('\t\t\t'+'in Elevator : '+str(person.inelevator)
                  +'\t'+'is Finished : '+str(person.isfinished))
        print('\n->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->->\n')    
    def run(self):
        self.building_elevator.register_customer()
        self.output()
        while len(self.building_elevator.register_list) + len(self.building_elevator.inelevator) != 0 :
            self.building_elevator.move()
            self.building_elevator.register_customer()
            self.building_elevator.cancel_customer()
            self.output()
            if self.building_elevator.cur_floor == self.building_h :
                print('The Elevator is on the Top Floor Now')
                self.building_elevator.direction = 'down'
        print('All customers Arrived at Destination')
    

a = building()
a.run()
```
