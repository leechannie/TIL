## 가설 세우기
```
1. 1월에 사망자가 가장 많을 것이다.
2. 41-50세 사이 가해자가 총 사고건수가 가장 많을 것이다
```

````python
import pandas as pd
import numpy as np
import seaborn as sns
````

````python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
# Window 의 한글 폰트 설정
plt.rc('font',family='Malgun Gothic')
# Mac 의 한글 폰트 설정
# plt.rc('font', family='AppleGothic') 
plt.rc('axes', unicode_minus=False)

# 폰트가 선명하게 보이기 위해
set_matplotlib_formats('retina')

# 그래프가 노트북 안에 보이게 하기 위해
%matplotlib inline
````

### 1. 데이터 로드 및 정보

````python
df = pd.read_csv('도로교통공단_가해운전자 연령층별 월별 교통사고 통계_20191231.csv',engine='python')
df.shape
````

````python
df.head(10)
````

````python
df.info
````

#### 가해자 나이 불명 제거

````python
df = df[df.가해자연령층 != "불명"]
df.tail(3)
````

### 2. 월별 사망자수 시각화

````python
df['사망자수'].min()
````

````python
df['사망자수'].max()
````

````python
plt.figure(figsize=(15,15))
df['사망자수'].value_counts().plot.bar()
````

````python
df.groupby(["월"]).mean()
````

````python
df.groupby(["월"]).max()
````

````python
sns.lmplot(
    data=df,
    x="월",
    y="사망자수",
)
````

````python
sns.lmplot(
    data=df,
     x="월",
     y="사망자수", 
     hue="가해자연령층"
)
````

#### 결론: 12월이 가장 사망자수가 많다 (가설 틀림)

### 3. 가해자연령별 사고건수 시각화

````python
plt.figure(figsize=(15,15))
sns.lmplot(
    data=df,
     x="월",
     y="사고건수", 
     hue="가해자연령층"
)
````

````python
plt.figure(figsize=(15,15))
sns.barplot(data=df, x="월", y="사고건수",hue="가해자연령층", ci=None)
````

````python
plt.figure(figsize=(15,15))
sns.barplot(data=df, x="가해자연령층", y="사고건수",hue="월", ci=None)
````

````python
sns.pointplot(data=df, x="가해자연령층", y="사고건수", ci=None)
````

#### 결론: 51-60세의 총 사고건수가 가장 많다 (가설 틀림)
